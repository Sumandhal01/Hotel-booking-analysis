## ***3. Data Wrangling***

### Data Wrangling Code

Handling Missing Data

````python
# Columns having missing values
hotel_bookingsdf.isnull().sum().sort_values(ascending=False)[:8]
````

````python
# There may be some cases where the customers booked hotel without the help of any agent or any company so in that case the values can be null.
# We will replace the null value with 0 for the column agent and company

hotel_bookingsdf[['company','agent']] = hotel_bookingsdf[['company','agent']].fillna(0)
````

````python
# Check if company and agent columns is cleaned
hotel_bookingsdf.isnull().sum().sort_values(ascending=False)[:8]
````

````python
hotel_bookings['children'].unique()
````

````python
# Children column has '0' as a value for the no children were present with the customers,so 'nan' must be missing values due to error in data
# We will replace the null value with the mean value for the children column

hotel_bookingsdf['children'].fillna(hotel_bookingsdf['children'].mean(), inplace=True)
````

````python
# Check if children column is cleaned
hotel_bookingsdf.isnull().sum().sort_values(ascending=False)[:8]
````

````python
# Next column with missing value is country, it represents the origin of the customers.
# Since it's datatype is string so we will replace the null value with others

hotel_bookingsdf['country'].fillna('others', inplace = True)
````

````python
# Check if all null values are cleaned
hotel_bookingsdf.isnull().sum().sort_values(ascending=False)[:8]
````

````python
# There are some rows with total number of adults, children or babies equal to zero that means no booking. So we will remove such rows.
# count of such rows

hotel_bookingsdf[hotel_bookingsdf['adults']+hotel_bookingsdf['children']+hotel_bookingsdf['babies']== 0].shape
````

````python
# Drop such rows
hotel_bookingsdf.drop(hotel_bookingsdf[hotel_bookingsdf['adults']+hotel_bookingsdf['children']+hotel_bookingsdf['babies']== 0].index, inplace= True)
````

````python
hotel_bookingsdf.shape
````

type casting the columns to appropriate datatypes

````python
# checking the datatypes of the column
hotel_bookingsdf.info()
````

````python
# As column company,agent,children had null values so there datatypes are not appropriate
# convert the datatype of column company,agent,children from float to int

hotel_bookingsdf[['children','company','agent']] = hotel_bookingsdf[['children','company','agent']].astype('int64')
````

````python
# As column reservation_status_date stores the date when the reservation is made So that column should be date-time
# convert reservation_status_date column to date-time

hotel_bookingsdf['reservation_status_date'] = pd.to_datetime(hotel_bookingsdf['reservation_status_date'], format ='%Y-%m-%d')
````

Adding columns


````python
# Adding some columns
hotel_bookingsdf['total_person'] = hotel_bookingsdf['adults'] + hotel_bookingsdf['children'] + hotel_bookingsdf['babies']
hotel_bookingsdf['kids'] = hotel_bookingsdf['children'] + hotel_bookingsdf['babies']
hotel_bookingsdf['stay_length'] = hotel_bookingsdf['stays_in_weekend_nights'] + hotel_bookingsdf['stays_in_week_nights']  # to analyse the stay lenght at hotel
# Creating 'guest_type' from variable 'total_person'
hotel_bookingsdf['guest_type']=np.where(hotel_bookingsdf['total_person']==1, 'single',np.where(hotel_bookingsdf['total_person']==2, 'couple', 'family'))
````

Handling Outliers

````python
# Changing name of some column for better understand
hotel_bookingsdf.rename(columns = {'adr':'average_daily_rate'},inplace = True)
````

````python
# Using Inter Quartile range in skew symmetric curve for removing outliers

# Outlier columns
outliers_columns=['lead_time','average_daily_rate']

# Capping dataset
for i in outliers_columns:
    #Findng IQR
    Q1=hotel_bookingsdf[i].quantile(0.25)
    Q3=hotel_bookingsdf[i].quantile(0.75)
    IQR=Q3-Q1

    # Defining lower and upper limit
    lower_limit =hotel_bookingsdf[i].quantile(0.25)-1.5*IQR
    upper_limit =hotel_bookingsdf[i].quantile(0.75)+1.5*IQR

    # Applying lower and upper limit to each variables
    hotel_bookingsdf.loc[(hotel_bookingsdf[i] > upper_limit),i] = upper_limit
    hotel_bookingsdf.loc[(hotel_bookingsdf[i] < lower_limit),i] = lower_limit
````
